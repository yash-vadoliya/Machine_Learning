1. Read the dataset and display the first 10 records

```python
import pandas as pd

# Read the dataset
df = pd.read_csv('world_literacy_data_2021_2023.csv')

# Display the first 10 records
print(df.head(10))
```

#2. Print all the column names

```python
# Print all column names
print(df.columns)
```

#3. Print the size (rows, cols) of the dataset

```python
# Print the size of the dataset
print(df.shape)
```

#4. Check for NULL values and replace them with the average values

```python
# Check for NULL values
print(df.isnull().sum())

# Replace NULL values with average values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Print the updated dataset
print(df.head(10))
```

5. Display the top 10 countries having highest GDP in 2023

```python
# Filter data for the year 2023 and sort by GDP
top_gdp_2023 = df[df['Year'] == 2023].sort_values(by='GDP', ascending=False).head(10)
print(top_gdp_2023[['Country', 'GDP' , 'Year']])
```

#6. Display the top 10 countries having highest Literacy rate in 2023

```python
# Filter data for the year 2023 and sort by Literacy Rate
top_literacy_2023 = df[df['Year'] == 2023].sort_values(by='Literacy Rate', ascending=False).head(10)
print(top_literacy_2023[['Country', 'Literacy Rate' , 'Year']])
```

#Plot the year-wise literacy rate of each country

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV file
df = pd.read_csv("data.csv")

# Ensure the data has the correct columns
required_columns = {'Year', 'Literacy Rate', 'Country'}
if not required_columns.issubset(df.columns):
    raise ValueError(f"CSV file must contain the following columns: {required_columns}")

# Create a figure and axis object
fig, ax = plt.subplots()

# Plotting each country's literacy rate over the years
for country in df['Country'].unique():
    country_data = df[df['Country'] == country]
    ax.plot(country_data['Year'], country_data['Literacy Rate'], label=country)

# Add labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Literacy Rate')
ax.set_title('Year-wise Literacy Rate of Each Country')

# Add a legend
ax.legend(title='Country')

# Display the plot
plt.show()
```
