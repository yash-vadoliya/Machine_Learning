```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("One.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Applying Kernal to an Image
# k1 = np.array([[0,0,0],[0,1,0],[0,0,0]])
k1 = np.array([[-2,-2,-2],[-2,1,-2],[-2,-2,-2]])
filter_img = cv2.filter2D(img,-1,k1)
plt.imshow(filter_img)
```

```python
# Blur Image
blur_img=cv2.blur(img,(5,5))
plt.imshow(blur_img)
```
