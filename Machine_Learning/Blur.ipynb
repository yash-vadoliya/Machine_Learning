```python
#  Blur Images
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("one.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

blurred_img1 = cv2.blur(img,(10,10))
blurred_img2 = cv2.GaussianBlur(img,(5,5),0)
blurred_img3 = cv2.medianBlur(img,3)

plt.subplot(2,2,1),plt.imshow(img),plt.title("Orignal Image")
plt.subplot(2,2,2),plt.imshow(blurred_img1),plt.title("blur Image")
plt.subplot(2,2,3),plt.imshow(blurred_img2),plt.title("GaussianBlur Image")
plt.subplot(2,2,4),plt.imshow(blurred_img3),plt.title("medianBlur Image")
```
